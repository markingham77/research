```python
# get tickers (which were downloaded from comapnymarketcap.com)
def get_data(tickers_csv_file='../data/top100_clothing_company_tickers.csv'):

    raw = pd.read_csv(tickers_csv_file)
    tickers = raw['Symbol']
    names = raw['Name']


    esg_data=[]
    for idx,ticker in enumerate(tickers):
        try:
            x=yesg.get_historic_esg(ticker.upper())
            y=x.iloc[-1]
            y.name=names[idx]
            esg_data.append(y)
        except:
            print(f'something wrong with {ticker}')    
    return pd.DataFrame(esg_data)        
```

```python
# let's try yfinance
import yfinance as yf
def get_esg_data(tickers):
    """
    retrieves ESG data using yfinance
    """
    # Retrieve Yahoo! Finance Sustainability Scores for each ticker
    for i in tickers:
        # print(i)
        i_y = yf.Ticker(i)
        try:
            if i_y.sustainability is not None:
                temp = pd.DataFrame.transpose(i_y.sustainability)
                temp['company_ticker'] = str(i_y.ticker)
                # print(temp)
                esg_data = esg_data.append(temp)
        except IndexError:
            pass
        
```
